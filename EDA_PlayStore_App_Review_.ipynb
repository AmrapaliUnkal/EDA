#except 'Varies with devices' lets standardise this column by converting Kb to mb

#We know that 1MB = 1024KB, to convert KB to MB, we must divide all the values which are in KB by 1024.

```python
def convert(size):
  if 'k' in size:
    return float(size[:-1])/1024
  elif 'M' in size:
    return float(size[:-1])
  else:
    return size
df['Size']=df['Size'].apply(lambda x: convert(x))
```

# 1. Which Category is most preffered by people?

```python
category_counts = df['Category'].value_counts()

# Get the category with the highest frequency
most_preferred_category = category_counts.idxmax()

print("The most preferred category by people is:", most_preferred_category)
```

```python
df['Category'].value_counts()
```

```python
#Lets represent it graphically
plt.figure(figsize = (10,12))
df['Category'].value_counts().plot(kind = 'barh', color = 'skyblue').invert_yaxis()
plt.title('Most prefered category')
```

Observations
